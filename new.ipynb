```python
import pandas as pd
import plt
```

```python
df = pd.read_excel("coffee.xlsx")
df.head()
```

```python
df.info()
```

```python
df["year"] = df["date"].dt.year
df.head()
```

```python
df["month"] = df["date"].dt.month_name()
df.head()
```

```python
df["day_name"] = df["date"].dt.day_name()
df.head()
```

```python
def time_check(time):
    if time.hour < 12:
        return "Morning"
    elif time.hour < 16:
        return "Afternoon"
    else:
        return "Evening"
```

```python
df["time_of_day"] = df ["datetime"].apply(time_check)
df.head()
```

```python
# How to drop/select columns
#To select multiple columns
df = df[["month","year","day"]]
```

```python
# How to drop columns you don't want
df = df.drop(["month_name","day_name","year"],axis = 1)
df.head()
```

```python
 # to rename a column
df = df.rename({"monthname" : "month"},axis=1)
df.head()
```

```python
df = df.drop(["date","datetime","card"],axis=1)
df.head()
```

```python
#DATA SUMMARY
#In Python we use the group by function for data summary. It also uses the aggregate function
monthly_revenue = df.groupby("month")["money"].sum()
monthly_revenue
```

```python
daily_sales = df.groupby("day_name")["day_name"].count()
daily_sales
```

```python
daily_rev = df.groupby("day_name")["money"].sum()
daily_rev
```

```python
busiest_day = df.groupby("time_of_day")["time_of_day"].count()
busiest_day
```

```python
popular_coffee = df.groupby("coffee_name")["coffee_name"].count()
popular_coffee
```

```python
payment_method = df.groupby("cash_type")["money"].count()
payment_method
```

```python
coffee_revenue = df.groupby("coffee_name")["money"].sum()
coffee_revenue
```

```python
coffee_revenue.plot(kind = "bar", color = "grey")
plt.ylabel("Total Revenue")
plt.title("Revenue per Coffee")
plt.xticks(rotation=45)
plt.show()
```
